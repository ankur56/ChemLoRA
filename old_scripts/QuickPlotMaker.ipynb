```python
import pandas as pd
```

```python
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
b3lyp_df = pd.read_json("outputs_gpt2_b3lyp.json")
b3lyp_df=b3lyp_df.dropna()
print(pearsonr(b3lyp_df['energy_true'], b3lyp_df['energy_out']))
plt.scatter(b3lyp_df['energy_true'], b3lyp_df['energy_out'])
plt.plot(b3lyp_df['energy_true'], b3lyp_df['energy_true'], color = 'black')
plt.axis('square')
```
